### Oh no it looks like we have two sets of columns in our df, lets take care of that too!! 

```python
# Since it appears to be the same we can comfortably drop the column
# Make some list to hold columns 

drop_list = ['Unnamed: 0']
x = []

for _id1 in df.columns:
    if '_id.1' in _id1:
        x.append(_id1)
    else:
        pass

for xs in x:
    df.drop(columns=xs,inplace=True)
    
    
for col in df.columns:
    if '_id' in col:
        df.drop(columns=col,inplace=True)

for cols in df.columns:
        if '.1' in cols:
            df.drop(columns=cols,inplace=True)
        else:
             pass


# Trust but verify our code 

df.drop(columns='Unnamed: 0',inplace=True)
```
